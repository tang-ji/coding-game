### The Goal
Capture more ghosts than the rival team.
### Rules

The game is played on a map **16001** units wide and **9001** units high. The coordinate **X=0, Y=0** is the top left pixel.

Each player controls a team of three **busters** – a **Hunter**, **Catcher**, and **Support**. The teams start out at opposite corners of the map, near their base. Ghosts are scattered across the map, and need to be trapped and brought back to a player's base to earn points. Each ghost inside the base or currently trapped by the Catcher is worth a point for the owner.

The map works as follows:

- There are always 2 teams in play.

- At the start of the game, each player is given a team id. This indicates on which corner their base is positioned. The top left corner (**X=0, Y=0**) is for team **0**. The bottom right corner (**X=16000, Y=9000**) is for team **1**.

- Fog makes it impossible to know the positions of ghosts and rival busters unless they are within **2200** units from one of your own busters.

- Each buster has a unique id. Each ghost has a unique id. Ghosts and busters with the same id are not correlated.

Busters work as follows:

- Each buster has a different role and can perform different actions :

 - The Hunter:
 
   He can **BUST** ghosts. He has a red outline in the viewer.
   
 - The Catcher:
 
   He can **TRAP** and **RELEASE** ghosts. He has a green outline in the viewer.
   
 - The Support:
 
   He can **STUN** other busters and use a **RADAR**. He has a blue outline in the viewer.
   
 - In addition all busters can MOVE.
 
**MOVE** followed by map coordinates will make the buster advance **800** units towards that point. The position will be rounded to the nearest integer.

**BUST** followed by a ghost's id will cause the Hunter to target that ghost with his proton stream, causing the ghost to lose **1** stamina point. He may do this if he is within a range of **1760** units around the target ghost but not closer than **900** units

**TRAP** followed by a ghost's id will cause the Catcher to try to suck that ghost into his ghost trap. If the ghost has **0** stamina points, and if the Catcher is within a range of **1760** units around the ghost but not closer than **900** units the action will suceed. Trapped ghosts are no longer visible on the map.

A Catcher may carry no more than **1** ghost at a time.

**RELEASE** will cause the Catcher to drop the ghost he is carrying back onto the map. If this is done within **1600** units of a corner of the map on which a base is positioned, the ghost is removed from play and secures a point for the base's owner.

**STUN** followed by a buster's id will make the Support shoot a bolt of lightning that will stun the target buster. A stunned buster cannot do anything for **10** turns. The stun ability must recharge for **20** turns before it can be used again. Being stunned will cause a Catcher to release any ghost he may be carrying.

The Support may only stun opponents within a **1760** unit radius.

**RADAR** will cause the Support to gain extended vision through the fog and will make it possible to see all entities within **4400** units around him for the next turn. Your Support can use this ability once per game.

Busters will perform their actions simultaneously but any effect the actions have, including movement, are applied at the end of the turn.

Ghosts work as follows:

- Ghosts remain stationary under two conditions:

 - They are occupying their starting position and the nearest buster is more than 2200 units away.

 - They are being busted.
 
- If a buster is closer than **2200** from a ghost, and this ghost is not being busted, the ghost will move away from the mean position of the near busters.

- If they aren't at their starting position, the ghosts will try to return directly to it. If this makes the ghost go closer than **2200** units to a buster, it will begin to flee again.

- The speed of the ghosts is **400** units.

- A ghost being carried by a Catcher will escape if he attempts to trap a different ghost.

- Ghost stamina does not regenerate.

The game ends once all ghosts have been captured or after the time limit of **250** turns.

The game state of every turn is given to you as a list of entities, each with an **id**, **position**, **type**, **role**, **state** and **value**.

The **type** will be:

- 0 for a buster from team 0.
- 1 for a buster from team 1.
- -1 for a ghost.

The **role** will be:

- 0 for the Hunter.
- 1 for the Catcher.
- 2 for the Support.
- -1 for a ghost.

The **state** will be:

- For busters:

 - 0: idle or moving buster.
 - 1: buster carrying a ghost.
 - 2: stunned buster.
 - 3: buster in the process of trapping a ghost.
 - 4: buster in the process of busting a ghost.
 
- For ghosts, it is equal to the ghost's current stamina.

The value will be:
- For busters:
 - If the buster is carrying or attempting to trap a ghost, that ghost's id.
 - If the buster is stunned the number of turns until he can move again.
- For ghosts it is the number the busters currently attempting to trap or bust it.
 

### Expert Rules
- The starting positions of ghosts and busters are symmetrical.
- Ghosts start moving 1 round after they spot you.
- If a ghost or buster attempts to leave the game area, it is snapped back to the nearest point within the game area.
- Using STUN on a buster who is already stunned will cause him to remain stunned for the next 10 rounds.
- If a buster carrying a ghost is stunned within a player's base, the ghost is removed from play and secures a point for the base's owner.

### Note
The program must first read within an infinite loop, read the contextual data from the standard input and provide to the standard output the desired instructions.

### Game Input

**Initialization input**

Line 1: an integer bustersPerPlayer for the amount of busters each team controls.

Line 2: an integer ghostCount for the amount of ghosts available on the map.

Line 3: an integer myTeamId for the team you are playing as.

**Input for one game turn**

Line 1: an integer entities for the amount of entities on the map that are visible to you.

Next entities lines: 6 space separated integers entityId, x, y, entityType, entityRole, state & value. Represents a buster or a ghost.

These are captured for you, and you can see how it is done at the top of your skeleton code, right after the infinite loop is created.

**Output for one game turn**
To submit a turn, you must make three calls to your language’s print statement function. Each call refers to a different buster, and the order matters.

- The first line is for your Hunter which can use MOVE or BUST.
- The second for your Catcher which can use MOVE, TRAP or RELEASE. 
- The third is for your Support which can use MOVE, STUN or RADAR. 

An example is provided within the skeleton code, just before the end of the infinite loop.

**One line for each buster**: one of the following:
- MOVE followed by two integers x and y
- BUST followed by one integer ghostId
- TRAP followed by one integer ghostId
- RELEASE
- STUN followed by one integer busterId
- RADAR

You may append text to your instructions, it will be displayed in the viewer.

### Constraints
**bustersPerPlayer** = 3

8 ≤ **ghostCount** ≤ 13

**stamina** in {3, 15, 40}

Response time per turn ≤ 100ms
